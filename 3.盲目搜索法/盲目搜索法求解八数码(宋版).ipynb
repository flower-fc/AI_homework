```python
import numpy as np
#矩阵进行四个方向的行动
def find_zero(num):

    x,y=np.where(num ==0 )

    return x[0],y[0]

def swap(num_data,dirction):

    x,y=find_zero(num_data)

    num=np.copy(num_data)

    if dirction == "l":

        if y ==0 :
            return num
        num[x][y] =num[x][y-1]
        num[x][y-1] = 0
        return num
    if dirction == "R":

        if y ==2 :
            return num
        num[x][y] =num[x][y+1]
        num[x][y+1] = 0
        return num
    if dirction == "u":

        if x ==0 :
            return num
        num[x][y] =num[x-1][y]
        num[x-1][y] = 0
        return num
    if dirction == "d":

        if x==2 :
            return num
        num[x][y] =num[x+1][y]
        num[x+1][y] = 0
        return num
#当前节点能产生的子类特征
def rect_search(num):
    c=[]
    dr=["l","u","R","d"]
    for i in dr:
        c.append(swap(num, i))
    return c

#判断子类是否都在三张表中
def judgement(nums,ps,nps,nss):
    repeat=0
    for i in nums:
        flag=0
        for j in ps:
            if (i==j).all():
                flag=1
        for k in nps:
            if (i==k).all():
                flag=1
        for l in nss:
            if (i==l).all():
                flag=1
        if(flag == 1):
            repeat=repeat+1
    return repeat
#寻找子类不重复的部分
def find_re(nums,ps,nps,nss):
    a=[]
    for i in nums:
        flag=0
        for j in ps:
            if (i==j).all():
                flag=1
        for k in nps:
            if (i==k).all():
                flag=1
        for l in nss:
            if (i==l).all():
                flag=1
        if(flag == 0):
            a.append(i)
    return a
#将寻找的不重复部分加入NPS表中
def add_new(nums,nps):
    for i in nums:
        nps.append(i)
    return nps
```

```python
def main():
    NPS=[] #定义NPS表 存储
    NSS=[]#定义NSS表 废除表，沿表内节点无法搜到结果
    PS=[]#定义PS表。表示转换过程
    CS=[]
    target=[]
    a = np.array([[2, 3, 1], [5, 0, 8], [4, 6, 7]])
    b= np.array([[2, 3, 0], [5, 6, 1], [4, 7, 8]])
    CS.append(a)
    PS.append(a)
    NPS.append(a)
    target.append(b)
    while(NPS!=[]):
       
        if((CS[0]==[])):
            print("查找失败")
            break
        if((CS[0]==target[0]).all()):
            for i in PS:
                print(i)
                print('-' * 50) 
            break
        #判断是否有无子状态
        if(judgement(rect_search(CS[0]),PS,NPS,NSS)==4):
            while((len(PS)!=0) and((CS[0]==PS[-1]).all())  ):
                NSS.append(CS[0])
                PS.pop()
                NPS.pop()
                CS[0]=NPS[-1]
            PS.append(CS[0])
        else:
            for i in  find_re(rect_search(CS[0]),PS,NPS,NSS):
                NPS.append(i)
            CS[0]=NPS[-1]
            PS.append(CS[0])
```
