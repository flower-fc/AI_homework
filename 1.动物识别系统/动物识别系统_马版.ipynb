```python
#con_list是本次的综合数据库,
#输入本次判断的综合数据库，并返回该数据库
def input_con():
    con_list=[]
    while 1:
        a= input()
        if a!='0':
            con_list.append(a)
        else:
            break
    con_dic = set(con_list)
    return con_dic
```

```python
#更新综合数据库里的数据
#返回经一次更新后的综合数据库和更新后的规则库
def update_con(con_dic,rules):
    for i in rules:
        flag = 1
        for j in i:
            if j not in con_dic:
                flag = 0
        if flag == 1:
            con_dic.add(rules[i])
            # del rules[i]
    return con_dic,rules
```

```python
#经过循环得到最终的综合数据库
#返回最终版本的综合数据库
def final_con(con_dic,rules):
    a=len(con_dic)
    fin_con_dic,rules = update_con(con_dic=con_dic,rules=rules)
    while 1:
        if a==len(fin_con_dic):
            break
        else:
            a=len(fin_con_dic)
            fin_con_dic,rules=update_con(con_dic=fin_con_dic,rules=rules)
    return fin_con_dic
```

```python
#查找是否综合数据库含有结论
#返回除去结论的综合数据库和结论
def veri(fin_con_dic,conclusion):
    # fin_con_dic=list(fin_con_dic)
    a=len(fin_con_dic)
    b=len(fin_con_dic)
    for i in fin_con_dic:
        if i in conclusion:
            b=b-1
            result = i
    # if a!=b-1:
    #     print(fin_con_dic)
    #     return fin_con_dic,'找不到结果'
    # else:
    return fin_con_dic,result
```
